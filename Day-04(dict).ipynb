# Count the frequency of each number in a list.

```python
a=[10,20,10,30,10,10,30,40,20,40,40,50]
b=dict()
for i in a:
    b[i]=b.get(i,0)+1
b
```

# Count the frequency of each character in a string.

```python
ch="Sujata Singh"
c=dict()
for i in ch:
    c[i]=c.get(i,0)+1
c
```

# Count the frequency of each word in a sentence.

```python
words="Apple Banana Apple Mango Mango Banana Grapes Mango"
a=words.split()
d=dict()
for i in a:
    d[i]=d.get(i,0)+1
d
```

# Find the key with the maximum value in a dictionary.

```python
d={
   "a":100,
   "b":250,
   "c":40,
   "d":200 
}
d    
```

```python
maximum=max(d,key=d.get)
maximum
```

# Find the key with the minimum value in a dictionary.

```python
minimum=min(d,key=d.get)
minimum
```

# Create a dictionary from two lists (keys and values).

```python
keys=['a','b','c']
Values=[1,2,3]
d={}
for i in range(len(keys)):
    d[keys[i]]=Values[i]
d
```

# Swap keys and values in a dictionary.

```python
swapped={}
for keys,Values in d.items():
    swapped[Values]=keys
swapped
```

# Find the total sum of all values in a dictionary.

```python
d={'a': 1, 'b': 2, 'c': 3}
total=sum(d.values())
total
```

# Find the average of all values in a dictionary.

```python
d={'a': 1, 'b': 2, 'c': 3}
avg=sum(d.values())/len(d)
avg
```
